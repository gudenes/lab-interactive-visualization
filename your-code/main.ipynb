# Interactive Visualization Lab

Complete the following set of exercises to solidify your knowledge of interactive visualization using Plotly, Cufflinks, and IPyWidgets.

```python
import pandas as pd
import plotly.plotly as py
import cufflinks as cf
from ipywidgets import interact

cf.go_offline()
```

```python
data2 = pd.read_excel('../data/Online Retail.xlsx')
```

```python
data = data2
```

## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
d1 = data[(data["Country"]!= "United Kingdom") & (data['InvoiceDate'] > '2011-04') & (data['InvoiceDate'] < '2011-5')]
```

```python
d1.head()
```

```python
d1 = d1.groupby(d1["Country"]).agg({"Quantity":"sum","Revenue":"sum"})
d1.head()
```

```python
d1 = d1.reset_index()
```

```python
d1.head()
```

```python
d1.iplot(kind='bar', x='Country', xTitle='Payment Method',
           yTitle='Total', title='Qtd and Revenue by Country')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
d2 = data[(data["Country"] == "France") 
     & (data['InvoiceDate'] > '2011-01-01') 
     & (data['InvoiceDate'] < '2011-05-31')]
```

```python
d2 = d2.groupby(d2["InvoiceDate"]).agg({"Quantity":"sum","Revenue":"sum"})
```

```python
d2 = d2.reset_index()
```

```python
d2.iplot(kind='line', x='InvoiceDate', xTitle='Period',
           yTitle='Total', title='Qtd and Revenue by Country')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
df3=data2[data2["Description"] == "PARTY BUNTING"]
df3=df3[["Quantity","UnitPrice","Country"]]
```

```python
df3 = df3.groupby(df3["Country"]).agg({"Quantity":"mean","UnitPrice":"mean"})
```

```python
df3 = df3.reset_index()
```

```python
df3.head()
```

```python
df3.iplot(x='Quantity', y='UnitPrice', categories='Country',
           xTitle='Quantity', yTitle='Unit Price',
           title='Charges vs. Tenure: One Year Contract, Credit Card Customers')
```

## 6. Create an interactive line chart showing quantity sold by day for the United Kingdom. Add drop-down boxes for Year and Month that allow you to filter the date range that appears in the chart.

```python
data['Year'] = pd.DatetimeIndex(data['InvoiceDate']).year
data['Month'] = pd.DatetimeIndex(data['InvoiceDate']).month
data['Day'] = pd.DatetimeIndex(data['InvoiceDate']).day
uk = data[data['Country']=='United Kingdom']
```

```python
uk.head()
```

```python
df6 = uk.groupby("Day", as_index=False).agg({"Quantity":"sum"})
```

```python
df6.iplot(kind='line', x='Day', xTitle='Days', color='black',
           yTitle='Quantity', title='Quantity Sold per Day')
```

```python
from ipywidgets import interact
```

```python
@interact(Selection=['Day', 'Year', 'Month'])

def linechart(Selection):
    data = uk.groupby(Selection, as_index=False).agg({"Quantity":"sum"})

    data.iplot(kind='line', x=Selection, xTitle='Days', color='black',
           yTitle='Quantity', title='Quantity Sold per ' + Selection.title())
    
```

## 7. Create an interactive scatter plot that plots number of invoices (x-axis) vs. number of customers (y-axis) and the plot points represent individual products. Add two sliders that control the x and y axis ranges.

```python
agg_func = {'InvoiceNo':'nunique',
            'Quantity':'sum',
            'UnitPrice':'mean',
            'Revenue':'sum',
            'CustomerID':'nunique'}

products = uk.groupby('Description').agg(agg_func)
```

```python
products.head()
```

```python
@interact(bins = (1, 1891))

def hist(bins):
    products.iplot(kind='scatter', x="InvoiceNo",y="CustomerID", bins=bins, title='Tenure Distribution')
```
